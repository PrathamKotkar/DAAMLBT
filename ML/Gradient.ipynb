```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: (x + 3)**2
```

```python
df = lambda x: 2 * (x + 3)
```

```python
cur_x = 2
rate = 0.01
precision = 0.000001
previous_step_size = 1
max_iters = 100
iters = 0
```

```python
iteration_list = []
x_values = []
f_values = []

while previous_step_size > precision and iters < max_iters:
    prev_x  = cur_x
    cur_x = cur_x - rate * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)
    iters = iters + 1
```

```python
iteration_list.append(iters)
x_values.append(cur_x)
f_values.append(f(cur_x))

print("Iteration", iters, "\nX value is", cur_x)

print("Local Minima: ", cur_x)
```

```python
x_plot = np.linspace(-10, 5, 400)
y_plot = f(x_plot)
```

```python
plt.figure(figsize = (12, 6))
```

```python
plt.subplot(1,2,1)
plt.plot(x_plot, y_plot, label = 'Function: f(x) = (x + 3) ^ 2')
plt.scatter(x_values, f_values, color = 'red', label = 'Gradient Descent Steps')
plt.title("Function and Gradient Descent Steps")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
```

```python
plt.subplot(1,2,2)
plt.plot(iteration_list, x_values, marker = 'o')
plt.title("Gradient Descent Steps")
plt.xlabel("Iteration")
plt.ylabel("X value")
plt.grid()

plt.tight_layout()
plt.show()
```
